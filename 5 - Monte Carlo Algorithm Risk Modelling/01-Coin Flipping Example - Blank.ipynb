# Coin Flipping Example

### Import Packages

```python
# Import packages
import numpy as np
import matplotlib.pyplot as plt
```

### Example 1 - Flip 20 coins 10 times

```python
# Set up variables
number_of_coins = 20
simulations = 10
p = 0.5

# Simulate the number of heads from flipping 20 coins 10 times
simulation_results = np.random.binomial(n=number_of_coins, p=p, size=simulations)
simulation_results
```

```python
# Test whether each simulation generated exactly 10 heads
sim_has_ten_heads = np.equal(simulation_results,10)
sim_has_ten_heads
```

```python
# Calculate the observed probability of getting 10 heads in any given simulation
ten_heads_probability = np.equal(simulation_results,10).mean()
ten_heads_probability
```

```python
# Calculate the observed probabilities of each number of heads
probabilities = [np.equal(simulation_results,i).mean() for i in range(number_of_coins)]
probabilities
```

```python
# Plot the observed probability distribution
plt.plot(list(range(number_of_coins)), probabilities, color='orange')
plt.xlabel('Number of Heads',fontsize=12)
plt.ylabel('Probability',fontsize=12)
plt.show()
```

### Example 2 - Flip 20 coins 10,000 times

```python
# Set up variables
number_of_coins = 20
simulations = 10000
p = 0.5

# Simulate the number of heads from flipping 20 coins 10,000 times
simulation_results = np.random.binomial(n=number_of_coins, p=p, size=simulations)
simulation_results

# Calculate the probabilities of each simulation result
probabilities = [np.equal(simulation_results,i).mean() for i in range(number_of_coins)]
probabilities

# Plot the probability distribution
plt.plot(list(range(number_of_coins)), probabilities, color='orange')
plt.xlabel('Number of Heads',fontsize=12)
plt.ylabel('Probability',fontsize=12)
plt.show()
```
